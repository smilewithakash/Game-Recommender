```python
df = df.dropna(subset = ['name'])
```

```python
df
```

```python
df.drop(['metacritic','esrb_rating', 'publishers', 'website', 'added_status_yet','added_status_owned', 'added_status_beaten', 'added_status_toplay', 'added_status_dropped', 'added_status_playing', 'playtime','tba', 'updated', 'playtime', 'achievements_count', 'ratings_count','released', 'id', 'slug','rating_top','reviews_count','game_series_count'], inplace = True, axis = 1)
```

```python
df['Published_Year'] = pd.DataFrame(Gdat).released
```

```python
df
```

```python
df.sort_values(by=['suggestions_count'] , ascending=False).head(20)
```

```python
newdf = df.copy(deep=True)
```

```python
newdf['platforms'] = newdf['platforms'].map(str).apply(lambda x:x.replace('||', ","))
newdf['developers'] = newdf['developers'].map(str).apply(lambda x:x.replace('||', ","))
newdf['genres'] = newdf['genres'].map(str).apply(lambda x:x.replace('||', ","))
```

```python
newdf['tags'] = newdf['rating'].map(str) + ', ' + newdf['suggestions_count'].map(str) + ', ' + newdf['platforms'].map(str) + ', ' + newdf['developers'].map(str) + ', ' + newdf['genres'].map(str) + ', ' + newdf['Published_Year'].map(str)
```

```python
newdf = newdf.loc[(newdf.rating != 0.0)]
```

```python
newdf.drop(['rating', 'suggestions_count', 'platforms', 'developers', 'genres' ,'Published_Year'], inplace = True, axis = 1)
```

```python
newdf['ratings'] = df['rating']
newdf['genres'] = df['genres']
```

```python
newdf.reset_index(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
```

```python
vector = cv.fit_transform(newdf['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(game):
    index = newdf[newdf['name'] == game].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(newdf.iloc[i[0]]['name'])
```

```python
newdf.sample(20)
```

```python
recommend('Dragon Ball Z: The Legacy of Goku')
```
